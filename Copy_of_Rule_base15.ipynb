```python
import pandas as pd
import io
import numpy as np
f = "https://raw.githubusercontent.com/AmujoeAdiday/MVPS/main/standardizedallstats_no_unname.csv"
file = pd.read_csv(f)
file
```

```python
years_trained = [year for year in range(2007,2022)]
file = file[file.Year.isin(years_trained)]
```

```python
file["Is_MVP_or_MVP_candidate"] = file["MVP"].apply(lambda x : 0 if x == 0 else 1)
file
```

```python
file = file.drop(["MVP"], axis =1 )
```

```python
overal_col_name = file.columns.values
overal_col_name
```

```python
def get_dictionary(col_name):
  return dict(zip(file.index, file[col_name]))
Player = get_dictionary("Player")
Team = get_dictionary("Tm")
Year = get_dictionary("Year")
```

```python
df = file.drop(["Player","GS","FG","FGA","FG%","eFG%","3P%","2P","2P%","FT","FT%","ORB","DRB","Year","Pts Won","Pts Max","Share","W","L","GB","Rank","PS/G","PA/G","Tm","Pos","Team"], axis = 1)
df["Is_MVP_or_MVP_candidate"] = df["Is_MVP_or_MVP_candidate"].astype(str)
df.info()
```

```python
features = df.columns.values[:-1]
```

```python
from sklearn.model_selection import train_test_split
from rulefit import RuleFit

X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
rf = RuleFit(max_rules=1,rfmode= "classify", model_type="r")
rf.fit(X_train, y_train,features )
```

```python
rules = rf.get_rules()
rules
```

```python
rules= rules.sort_values(by="importance", ascending=False)
pd.set_option("display.max_colwidth", 999)
rules.head(10)
```
